#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x & y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames)
```

    - Basic Logic: OLS is a linear regression model that fits a line to minimize the sum of squared differences between observed and predicted values.
    - Assumptions: Assumes a linear relationship between features and the target variable, and residuals are normally distributed.
    - Pros: Easy to interpret, and fast to train.
    - Cons: Not well-suited for non-linear relationships or high multicollinearity among features.
    - Relevance: Useful as a baseline to understand the linear dependencies in predicting diabetes rates.
- Model 2: Ridge Regression
    - Basic Logic: Ridge regression is a linear regression model with L2 regularization, which penalizes large coefficients to prevent overfitting.
    - Assumptions: Assumes linear relationships and addresses multicollinearity by shrinking coefficients.
    - Pros: Handles multicollinearity better than OLS; reduces overfitting.
    - Cons: Still assumes linearity, and parameter tuning is required.
    - Relevance: Useful in datasets where there might be a lot of correlated features.
- Model 3: Lasso Regression
    - Basic Logic: Lasso regression is a linear model with L1 regularization, which penalizes the absolute value of coefficients. It also performs feature selection by shrinking some coefficients to zero.
    - Assumptions: Assumes linear relationships and addresses feature selection effectively.
    - Pros: Helps with feature selection by zeroing out less important features; useful for reducing model complexity.
    - Cons: May require a low alpha to converge, especially with correlated features.
    - Relevance: Suitable for feature selection and avoiding overfitting when there are many predictors.

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

**YOUR ANSWER HERE**
- The bias-variance tradeoff refers to the balance between two types of errors in model prediction: bias and variance. Bias is the error introduced by approximating a real-world problem, which may be complex, with a simpler model. High bias can lead to underfitting, where the model fails to capture the underlying patterns in the data. Variance, on the other hand, is the error due to the models sensitivity to small fluctuations in the training set, leading to overfitting and poor generalizability.

- In our problem of predicting diabetes rates, managing the bias-variance tradeoff is critical. If our model is too simple (high bias), it will fail to capture meaningful relationships between the predictors and diabetes prevalence. On the other hand, if the model is too complex (high variance), it might perform very well on the training data but fail to generalize to new data, such as the validation or test sets. By carefully choosing model complexity and using techniques such as Ridge or Lasso regularization, we can balance the bias and variance to make the model perform well on unseen data, thus aiding policymakers in making informed decisions.
